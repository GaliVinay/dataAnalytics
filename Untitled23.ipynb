```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Open raster file
with rasterio.open("V:\\P5_PAN_CD_N16_000_E080_000_30m\\streams.tif") as dataset:
    band1 = dataset.read(1)  # Read first band

# Display the raster
plt.imshow(band1, cmap="plasma")
plt.colorbar(label="Pixel Values")
plt.title("Raster Visualization")
plt.show()
```

```python
from osgeo import gdal

input_raster = "V:\\P5_PAN_CD_N16_000_E080_000_30m\\streams.tif"
output_raster = "reprojected.tif"
target_crs = "EPSG:4326"  # New projection (WGS84)

gdal.Warp(output_raster, input_raster, dstSRS=target_crs)
```

```python
plt.imshow(output_raster, cmap="plasma")
plt.colorbar(label="Pixel Values")
plt.title("Raster Visualization")
plt.show()
```
